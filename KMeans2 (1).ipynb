```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\Abel\Downloads\crime_data.csv')
```

```python
sns.pairplot(df)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
scaler=StandardScaler()
df_norm=scaler.fit_transform(df.iloc[:,1:])
```

```python
#creating cluster
model=KMeans(n_clusters=4).fit(df_norm)
model.labels_
```

```python
df['clust']=model.labels_
```

```python
df.head()
```

```python
df.groupby('clust').mean()
```
